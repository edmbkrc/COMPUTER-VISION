```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist # hand writing recognition
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images = train_images/255.0
test_images = test_images/255.0
```

```python
plt.imshow(train_images[89], cmap="gray");
```

```python
# Convolutional Neural Networks---CNNs---Image Classification
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, InputLayer, Reshape, MaxPooling2D, Flatten, Dropout, BatchNormalization
```

```python
model=Sequential() 
model.add(InputLayer(input_shape=(28,28))) 
model.add(Reshape(target_shape=(28,28,1)))
model.add(Conv2D(filters=12,kernel_size=(3,3),activation='relu')) 
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(10)) # 10 fakli cevap classification 0-9 a kadar olan rakamlar
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, validation_split=.10, epochs=10)
```

```python
loss, accuracy = model.evaluate(test_images, test_labels)
print(f"ACCURACY: {round(accuracy,3)}\nLOSS: {round(loss,3)}")
```

```python
model.save('model_mnist.h5')
```

### Increasing success rate on deep learning
1. increase layer count
2. increase neuron count
3. add dropout and prevent overfitting
4. BatchNormalization

```python
model=Sequential() 
model.add(InputLayer(input_shape=(28,28))) 
model.add(Reshape(target_shape=(28,28,1)))
model.add(Conv2D(filters=12,kernel_size=(3,3),activation='relu')) 
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(BatchNormalization()) # make layers optimize it
model.add(Dense(60))
model.add(BatchNormalization()) # make layers optimize it
model.add(Dropout(.25)) # drop out unnecessary datas
model.add(BatchNormalization()) # make layers optimize it
model.add(Dense(10)) # 10 fakli cevap classification 0-9 a kadar olan rakamlar
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, validation_split=.10, epochs=10)
```

```python
loss, accuracy = model.evaluate(test_images, test_labels)
print(f"ACCURACY: {round(accuracy,3)}\nLOSS: {round(loss,3)}")
```
