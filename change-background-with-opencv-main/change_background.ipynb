ARKA PLAN DEĞİŞTİRME

OpenCV, cv2_imshow, numpy ve matplotlib kütüphanelerini içe aktarır. Bu kütüphaneler, görüntü işleme, görüntüleri gösterme ve veri görselleştirme işlemlerinde kullanılır.

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```



* Bu kod parçası, OpenCV kütüphanesini kullanarak "/content/foto.jpg" dosyasını okuyan ve bu resmi ekranda gösteren bir işlevi içeriyor.
* cv2.imread("/content/foto.jpg") ifadesi, "/content/foto.jpg" yolundaki bir resmi okumak için OpenCV kütüphanesinin imread fonksiyonunu kullanır.
* Okunan resim, foto adlı bir değişkene atanır.
* cv2_imshow işlevi, foto adlı değişken içindeki resmi ekranda göstermek için kullanılır. Bu işlev, görüntüyü açar ve bir pencerede görüntülemek için OpenCV kütüphanesini kullanır.








```python
foto = cv2.imread("/content/foto.jpg")
cv2_imshow(foto)
```

Aynı işlemler "/content/arka_plan.jpg" için de uygulanır.

```python
arka_plan = cv2.imread("/content/arka_plan.jpg")

cv2_imshow(arka_plan)
```

* cv2.resize(arka_plan, (foto.shape[1], foto.shape[0])) ifadesi, arka_plan adlı görüntüyü foto adlı başka bir görüntünün boyutuna yeniden boyutlandırmak için OpenCV kütüphanesinin resize fonksiyonunu kullanır. Yeniden boyutlandırma, foto.shape[1] ve foto.shape[0] değerleri kullanılarak gerçekleştirilir. Bu değerler, foto görüntüsünün genişliği ve yüksekliği olarak alınır.
* Yeniden boyutlandırılmış görüntü, arka_plan adlı bir değişkene atanır.
* cv2_imshow işlevi, arka_plan adlı değişken içindeki yeniden boyutlandırılmış görüntüyü ekranda göstermek için kullanılır. Bu işlev, görüntüyü açar ve bir pencerede görüntülemek için OpenCV kütüphanesini kullanır.

```python
arka_plan = cv2.resize(arka_plan, (foto.shape[1], foto.shape[0]))
cv2_imshow(arka_plan)
```




> Bu kod parçası, foto adlı görüntünün türünü, boyutunu ve daha sonra bu görüntünün bir kopyasını oluşturarak bu kopyayı ekranda gösteriyor.






```python
print('Bu resim:', type(foto), "türünde ve " , foto.shape, "boyutlarındadır.")

image_copy = np.copy(foto)
plt.imshow(image_copy);
```

* cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB) ifadesi, image_copy adlı görüntünün renk uzayını BGR'den RGB'ye dönüştürmek için OpenCV kütüphanesinin cvtColor fonksiyonunu kullanır. BGR, Blue-Green-Red, yani mavi-yeşil-kırmızı renk kanallarının sırasını ifade eder. Bu ifade, görüntünün renklerini doğru şekilde yorumlamak ve göstermek için dönüştürme işlemi yapar.
* Dönüştürülen görüntü, image_copy adlı değişkene atanır.
* plt.imshow(image_copy) ifadesi, image_copy adlı görüntüyü ekranda göstermek için Matplotlib kütüphanesinin imshow fonksiyonunu kullanır. Bu fonksiyon, görüntüyü bir figürde görüntülemek için kullanılır. Görüntünün RGB renk uzayında olduğu varsayılarak doğru şekilde gösterilir.

```python
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy);
```



> lower_blue = np.array([0,0,140]) ifadesi, mavi renk aralığının alt sınırını belirlemek için bir NumPy dizisi kullanır. Dizi içindeki [0, 0, 140] değerleri, mavi renk bileşeninin (Blue) alt sınırının [0, 0, 140] olduğunu belirtir. Bu değerler, BGR renk uzayında mavi renk bileşeninin en düşük değerini temsil eder.

> upper_blue = np.array([250,250,255]) ifadesi, mavi renk aralığının üst sınırını belirlemek için bir NumPy dizisi kullanır. Dizi içindeki [250, 250, 255] değerleri, mavi renk bileşeninin (Blue) üst sınırının [250, 250, 255] olduğunu belirtir. Bu değerler, BGR renk uzayında mavi renk bileşeninin en yüksek değerini temsil eder.

> Bu renk aralığı belirlemesi, genellikle bir görüntüde belirli bir renk aralığına sahip nesneleri veya bölgeleri tespit etmek için kullanılır. Örneğin, belirli bir mavi renge sahip nesneleri ayırt etmek veya izole etmek için bu renk aralığını kullanabilirsiniz.










```python
lower_blue = np.array([0,0,140])
upper_blue = np.array([250,250,255])
```

> cv2.inRange(image_copy, lower_blue, upper_blue) ifadesi, image_copy adlı görüntüde lower_blue ve upper_blue aralığına uyan pikselleri içeren bir maske oluşturmak için OpenCV kütüphanesinin inRange fonksiyonunu kullanır. Bu fonksiyon, görüntünün belirli bir renk aralığına dahil olan pikselleri belirlemek için kullanılır.

> Oluşturulan maske, mask adlı bir değişkene atanır.

> plt.imshow(mask, cmap='gray') ifadesi, mask adlı maskeyi ekranda göstermek için Matplotlib kütüphanesinin imshow fonksiyonunu kullanır. Bu fonksiyon, gri tonlama (cmap='gray') ile maskeyi görselleştirir. Maske, renkli bir görüntü yerine siyah-beyaz bir görüntü olarak gösterilir.

```python
mask = cv2.inRange(image_copy, lower_blue, upper_blue)
plt.imshow(mask, cmap='gray');
```

> masked_image = np.copy(image_copy) ifadesi, image_copy adlı görüntünün bir kopyasını masked_image adlı başka bir değişkene atar. Bu işlem, orijinal görüntüyü korumak ve üzerinde değişiklik yapmak istendiğinde kullanışlıdır. np.copy fonksiyonu, NumPy kütüphanesinin bir parçasıdır ve bir diziyi kopyalamak için kullanılır.

> masked_image[mask != 0] = [0, 0, 0] ifadesi, mask adlı maskeyi kullanarak masked_image adlı görüntüde maske dışında kalan pikselleri siyah renk (BGR [0, 0, 0]) ile değiştirir. Bu işlem, maske içinde olmayan bölgeleri (mask != 0) siyah renge dönüştürerek bu bölgeleri gizler.

```python
masked_image = np.copy(image_copy)
masked_image[mask != 0] = [0, 0, 0]
plt.imshow(masked_image);
```

* background_image = cv2.cvtColor(arka_plan, cv2.COLOR_BGR2RGB) ifadesi, arka_plan adlı arka plan görüntüsünün renk uzayını BGR'den RGB'ye dönüştürmek için OpenCV kütüphanesinin cvtColor fonksiyonunu kullanır. BGR, Blue-Green-Red, yani mavi-yeşil-kırmızı renk kanallarının sırasını ifade eder. Bu ifade, görüntünün renklerini doğru şekilde yorumlamak ve göstermek için dönüştürme işlemi yapar.

* crop_background = background_image[0:263, 0:739] ifadesi, background_image adlı görüntünün belirli bir bölümünü crop_background adlı bir değişkene atar. Bu işlem, background_image görüntüsünün 0'dan 263'e ve 0'dan 739'a kadar olan satır ve sütun aralığını seçer.

* crop_background[mask == 0] = [0,0,0] ifadesi, mask adlı maske ile uyumsuz olan pikselleri crop_background adlı görüntüde siyah renk (RGB [0, 0, 0]) ile değiştirir. Bu işlem, maske içinde olmayan bölgeleri (mask == 0) siyah renge dönüştürerek bu bölgeleri gizler.

```python
background_image = cv2.cvtColor(arka_plan, cv2.COLOR_BGR2RGB)
crop_background = background_image[0:263, 0:739]

crop_background[mask == 0] = [0,0,0]
plt.imshow(crop_background);
```

> Bu kod parçası, masked_image ve crop_background görüntülerini toplayarak birleştirir ve son_resim adlı bir değişkene atar. Son olarak, bu birleştirilmiş görüntüyü ekranda gösterir.

```python
son_resim = masked_image + crop_background
plt.imshow(son_resim);
```

TRANSPARAN GÖRÜNTÜ OLUŞTURMA






```python
# Arka plan resmini yükle
background = cv2.imread("/content/arka_plan.jpg")

# Fotoğrafı yükle
image = cv2.imread("/content/foto.jpg")

# Arka plan ve fotoğrafın boyutlarını eşitle
background = cv2.resize(background, (image.shape[1], image.shape[0]))

# Arka planı BGR'den RGB'ye dönüştür
background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)

# Fotoğrafı alfa kanalıyla birleştir
alpha = 0.2  # Arka plana uygulanacak alfa değeri (0.0 - 1.0 arasında)
result = cv2.addWeighted(background, alpha, image, 1 - alpha, 0)

# Sonucu kaydet
output_path = "/content/sonuc.jpg"
cv2.imwrite(output_path, cv2.cvtColor(result, cv2.COLOR_RGB2BGR))

# Sonucu görüntüle
output_image = cv2.imread(output_path)
cv2_imshow(output_image)
```
